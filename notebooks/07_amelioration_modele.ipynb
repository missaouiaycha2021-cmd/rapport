```python
import numpy as np
import joblib

model = joblib.load('models/isolation_forest_model.pkl')
scaler = joblib.load('models/scaler.pkl')
X_windows = np.load('data/processed/windows/X_windows.npy')

X_sample = X_windows[:3000]
X_flat = X_sample.reshape(X_sample.shape[0], -1)
X_scaled = scaler.transform(X_flat)

scores = model.decision_function(X_scaled)

# On propose 3 seuils différents
print("Scores statistiques :")
print(f"Min score     : {scores.min():.4f}")
print(f"Mean score    : {scores.mean():.4f}")
print(f"Max score     : {scores.max():.4f}")

print("\nSeuils recommandés :")
print(f"Seuil conservateur (détecte ~5%)  : {np.percentile(scores, 95):.4f}")
print(f"Seuil équilibré (détecte ~10-15%) : {np.percentile(scores, 87):.4f}")
print(f"Seuil sensible (détecte ~20%)     : {np.percentile(scores, 80):.4f}")
```
