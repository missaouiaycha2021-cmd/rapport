```python
import numpy as np
import os
import pandas as pd

print("=== CHARGEMENT DES LABELS D'ANOMALIES ===\n")

DATA_PATH = 'data/raw/SMD/'
WINDOWS_PATH = 'data/processed/windows/'

# Charger les fenêtres et les infos machines
X_windows = np.load(os.path.join(WINDOWS_PATH, 'X_windows.npy'))
machine_info = np.load(os.path.join(WINDOWS_PATH, 'machine_info.npy'))

print(f"X_windows shape : {X_windows.shape}")
print(f"Nombre de machines : {len(np.unique(machine_info))}\n")

# Charger les labels d'anomalies (test_label)
label_path = os.path.join(DATA_PATH, 'test_label')
label_files = sorted([f for f in os.listdir(label_path) if f.endswith('.txt')])

print(f"Nombre de fichiers de labels trouvés : {len(label_files)}")

# Créer un dictionnaire machine → labels
machine_labels = {}

for label_file in label_files:
    machine_name = label_file.replace('.txt', '')
    full_path = os.path.join(label_path, label_file)
    
    labels = np.loadtxt(full_path, dtype=int)
    machine_labels[machine_name] = labels

print("Labels chargés pour les machines de test.\n")

# Pour l'instant, on va créer des labels "0" (normal) pour toutes nos fenêtres d'entraînement
# (On utilisera les labels de test plus tard pour l'évaluation)

y_windows = np.zeros(X_windows.shape[0], dtype=int)

print("Labels créés (toutes fenêtres marquées comme normales pour l'entraînement)")
print(f"Shape de y_windows : {y_windows.shape}")
print(f"Nombre d'anomalies pour l'instant : {np.sum(y_windows)} (0)")

# Sauvegarde
np.save(os.path.join(WINDOWS_PATH, 'y_windows.npy'), y_windows)

print("\n✓ Labels sauvegardés dans data/processed/windows/y_windows.npy")

# Statistiques
print("\nRésumé :")
print(f"Fenêtres totales     : {X_windows.shape[0]:,}")
print(f"Fenêtres normales    : {np.sum(y_windows == 0):,}")
print(f"Fenêtres anomalies   : {np.sum(y_windows == 1):,}")

print("\nProchaine étape : Construction du modèle (LSTM Autoencoder)")
```
