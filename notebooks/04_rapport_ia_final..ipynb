```python
### Objectif de la partie IA

Développer un système intelligent capable de :
- Surveiller en temps réel les métriques des serveurs et VMs (CPU, RAM, stockage, réseau)
- Détecter les anomalies et risques de pannes avant quils ne surviennent
- Anticiper les saturations de ressources
- Fournir des alertes exploitables pour la maintenance prédictive
```

```python
### 1. Dataset et Exploration

Nous avons utilisé le **Server Machine Dataset (SMD)**, un dataset réel contenant 28 machines avec 38 métriques anonymisées.

Analyse des features principales :
- `feature_5` : fortement corrélée à l’utilisation RAM
- `feature_0` à `feature_3` : métriques liées à l’utilisation CPU
- `feature_22` : probablement liée à l’utilisation du disque
```

```python
### 2. Augmentation des Données

Pour améliorer la robustesse du modèle, nous avons généré 4 versions des données pour chaque machine :
- Version originale
- Version avec bruit gaussien
- Version avec scaling
- Version avec décalage temporel + bruit

**Total : 112 fichiers augmentés**
```

```python
### 3. Préparation des Données

Création de **fenêtres temporelles** (Sliding Windows) :
- Taille de fenêtre : 60 minutes
- Stride : 10 minutes
- Total de fenêtres créées : environ 70 687

Cela permet au modèle d’analyser l’évolution temporelle des métriques.
```

```python
### 4. Modèle de Détection d'Anomalies

Nous avons implémenté un modèle basé sur **Isolation Forest** combiné à des règles métier simples.

Un script `predict_anomaly.py` a été développé pour permettre l’intégration avec l’infrastructure.
```
